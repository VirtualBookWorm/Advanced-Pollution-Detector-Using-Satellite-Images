```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataSet = pd.read_csv('Data/pm25.csv')
dataSet
```

```python
dataSet.plot(x='Months',y='AQI_Value',style='.')
plt.title('AQI data of last 5 years')
plt.xlabel('Months from 01-01-2015 to 01-09-2019')
plt.ylabel('PM2.5 Level')
plt.show()
```

```python
dataSet.plot(x='Months',y='AQI_Value',style='-')
plt.title('AQI data of last 5 years')
plt.xlabel('Months from 01-01-2015 to 01-09-2019')
plt.ylabel('PM2.5 Level')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x = dataSet.iloc[:,:-1].values
y = dataSet.iloc[:,1].values
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.18)
```

```python
from sklearn.linear_model import LinearRegression
regressor  =LinearRegression()
regressor.fit(x_train,y_train)
```

```python
plt.scatter(x=x_train,y=y_train)
```

```python
print(regressor.intercept_)
print(regressor.coef_)
```

```python
y_pred = regressor.predict(x_test)
print(x_pred)
```

```python
df = pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
df
```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score
print('Mean Squared Error : ', mean_squared_error(y_test, y_pred))
print('Root means squared error : ', np.sqrt(mean_squared_error(y_test, y_pred)))
print('r_2 statistic (Accuracy): %.2f' %r2_score(y_test, y_pred))
```
